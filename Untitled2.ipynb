```python
import pandas as pd

df = pd.read_csv("/content/covid19_patient_symptoms_diagnosis.csv")
df.head()
```

```python
import sqlite3

# Create SQLite database
conn = sqlite3.connect("covid_data.db")
```

```python
df.to_sql("covid_patients", conn, if_exists="replace", index=False)
```

```python
query = """
SELECT gender, COUNT(*) AS total_patients
FROM covid_patients
GROUP BY gender;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT AVG(age) AS avg_age
FROM covid_patients
WHERE fever = 1;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS high_risk_patients
FROM covid_patients
WHERE fever + dry_cough + fatigue >= 3;
"""

pd.read_sql(query, conn)
```

```python
conn.close()
```
